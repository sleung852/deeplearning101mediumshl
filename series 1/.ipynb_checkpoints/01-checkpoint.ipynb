```python
aapl = yf.download('AAPL', 
                      start='2020-01-01', 
                      end='2020-05-31', 
                      progress=False)
aapl.head()
```

```python
aapl_df = aapl.loc['2020-05-01': '2020-05-07':,'Adj Close'].copy()
```

```python
aapl_df.plot(style='+', color='r')
plt.title('Stock Price of AAPL')
plt.ylabel('Adj Close')
plt.savefig('aapl.png')
```

```python
print(aapl_df)
```

```python
m0 = 30
c0 = 3
```

```python
y_pred = []
x = np.arange(1,6)
y_pred = m0 * x + c0
```

```python
loss = math.sqrt(sum((aapl_df - y_pred)**2))
```

```python
print("Total Loss: {}".format(loss))
```

```python
dm = - math.sqrt(sum((aapl_df - y_pred)* x) * 2 )
```

```python
dc = - math.sqrt(sum((aapl_df - y_pred)) * 2 )
```

```python
def train(m0, c0, epoch = 10000, learning_rate = 0.01):
    m = m0
    c = c0
    
    ms = [m0]
    cs = [c0]
    
    for i in range(epoch):
        if i % 25 == 0:
            print("Epoch {}".format(i))
        #model
        y_pred = []
        x = np.arange(1,6)
        y_pred = m * x + c
        #loss
        loss = math.sqrt(sum((aapl_df - y_pred)**2))
        if i % 25 == 0:
            print("Loss: {}".format(loss))
        #derivatives
        dm = sum((2 * y_pred - 2 * aapl_df) * x)
        dc = sum((2 * y_pred - 2 * aapl_df))
        #update
        m -= learning_rate * dm
        c -= learning_rate * dc
        ms.append(m)
        cs.append(c)
    
    return m, c, [ms, cs]
```

```python
m_new, c_new, record = train(1, 300)
```

```python
y_new = np.arange(1,6) * m_new + c_new
ax = plt.figure().gca()
ax.xaxis.set_major_locator(MaxNLocator(integer=True))

plt.plot(np.arange(1,6), y_new, color = 'g')
plt.plot(np.arange(1,6), aapl_df.values, 'r+')
plt.title('Stock Price of AAPL')
plt.ylabel('Adj Close')
plt.xlabel('Trading Day')
```

```python
plt.style.use('seaborn-pastel');

fig = plt.figure();
ax = plt.axes(xlim=(1, 5), ylim=(288, 304));
ax.xaxis.set_major_locator(MaxNLocator(integer=True));
line, = ax.plot([], [], lw=3);

ax.plot(np.arange(1,6), aapl_df.values, 'r+');

ax.set_ylabel('Adj Close');
ax.set_xlabel('Trading Day');

# generate the y_hats
y_records = []
ms, cs = record
for m, c in zip(ms, cs):
    y_records.append(m * np.arange(1,6) + c)

def init():
    line.set_data([], [])
    return line,
def animate(i):
    ax.set_title('Stock Price of AAPL | {}th Adjustment'.format(i*25))
    x = np.arange(1,6)
    y = y_records[i]
    line.set_data(x, y)
    return line,

anim = FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=12, blit=True)
```
